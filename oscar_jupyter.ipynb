```python
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split, StratifiedKFold

# 0. Setup
RANDOM_SEED = 42

# 1. Target Creation
# If Player_1 is the Winner, target is 1, else 0
y = (df['Winner'] == df['Player_1']).astype(int)

# 2. Feature Selection
numerical_cols = ['Rank_1', 'Rank_2', 'Pts_1', 'Pts_2', 'Odd_1', 'Odd_2', 'Best of']
categorical_cols = ['Series', 'Court', 'Surface', 'Round']

# Select features X
X = df[numerical_cols + categorical_cols].copy()

# Replace -1 with NaN in numerical columns to allow SimpleImputer to handle them
X[numerical_cols] = X[numerical_cols].replace(-1, np.nan)

# 3. Data Splitting (Prevent Leakage)
# Split BEFORE preprocessing
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=RANDOM_SEED, stratify=y
)

print(f"Training shape: {X_train.shape}")
print(f"Testing shape: {X_test.shape}")

# 4. Preprocessing Pipeline

# Numerical Pipeline: Impute missing values (median) -> Scale
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

# Categorical Pipeline: OneHotEncode
categorical_transformer = OneHotEncoder(handle_unknown='ignore', sparse_output=False)

# Combine transformers
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numerical_cols),
        ('cat', categorical_transformer, categorical_cols)
    ]
)

# Full Pipeline
pipeline = Pipeline(steps=[('preprocessor', preprocessor)])

# 5. Execution (Fit on Train ONLY)
# Fit the pipeline on training data
pipeline.fit(X_train)

# Transform both Train and Test
X_train_processed = pipeline.transform(X_train)
X_test_processed = pipeline.transform(X_test)

# Convert back to DataFrame for better visualization
# Get feature names from OHE
ohe_feature_names = preprocessor.named_transformers_['cat'].get_feature_names_out(categorical_cols)
all_feature_names = numerical_cols + list(ohe_feature_names)

X_train_final = pd.DataFrame(X_train_processed, columns=all_feature_names)
X_test_final = pd.DataFrame(X_test_processed, columns=all_feature_names)

# 6. Cross-Validation Setup
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=RANDOM_SEED)

print("Shape of X_train_final:", X_train_final.shape)
print("Shape of X_test_final:", X_test_final.shape)
X_train_final.head()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
from sklearn.model_selection import cross_val_score

# 7. Model Building (Logistic Regression)

# Create a full pipeline with the classifier
full_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', LogisticRegression(random_state=RANDOM_SEED, max_iter=1000))
])

# Fit the model on the training data
full_pipeline.fit(X_train, y_train)

# Predict on the test data
y_pred = full_pipeline.predict(X_test)

# Evaluate
print("Test Set Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))

# 8. Cross-Validation
print("Running Cross-Validation (this may take a moment)...")
cv_scores = cross_val_score(full_pipeline, X_train, y_train, cv=cv, scoring='accuracy')
print(f"Mean CV Accuracy: {cv_scores.mean():.4f} (+/- {cv_scores.std():.4f})")
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt
from sklearn.model_selection import cross_val_score

# Build Decision Tree pipeline (re-uses `preprocessor` defined earlier)
dt_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', DecisionTreeClassifier(random_state=RANDOM_SEED))
])

# Fit on train, predict on test
dt_pipeline.fit(X_train, y_train)
y_pred_dt = dt_pipeline.predict(X_test)

print('Decision Tree Test Accuracy:', accuracy_score(y_test, y_pred_dt))
print('\nClassification Report:\n', classification_report(y_test, y_pred_dt))

# Cross-Validation on training data
cv_scores_dt = cross_val_score(dt_pipeline, X_train, y_train, cv=cv, scoring='accuracy')
print(f'Mean CV Accuracy (Decision Tree): {cv_scores_dt.mean():.4f} (+/- {cv_scores_dt.std():.4f})')

# Get feature names for plotting (after preprocessor fit)
ohe_feature_names = dt_pipeline.named_steps['preprocessor'].named_transformers_['cat'].get_feature_names_out(categorical_cols)
feature_names = numerical_cols + list(ohe_feature_names)

# Plot a shallow version of the tree for readability
plt.figure(figsize=(20,10))
plot_tree(dt_pipeline.named_steps['classifier'], feature_names=feature_names, class_names=['Player_2','Player_1'], filled=True, max_depth=3)
plt.show()
```
